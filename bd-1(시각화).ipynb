```python
import pandas as pd
df_1 = pd.DataFrame(['1', '2', '3'], columns = ['A'])
df_2 = pd.DataFrame(['4', '5', '6', '7'], columns = ['B'])

# inner 로 하면 데이터가 적은 A를 기준으로 합쳐짐(B 가 잘림)
# outer 로 하면 데이터가 많은 B를 기준으로 합쳐짐 
# how를 설정 안 해주면 inner로 기본값
df = df_1.join(df_2, how = 'outer')
print(df_1)
print(df_2)
df
```

```python
# matplotlib 플롯(그래프)를 그릴 때 사용하는 시각화 라이브러리
import matplotlib as mpl
import matplotlib.pyplot as plt # 서브패키지 - 그려주는
import numpy as np

#%matplotlib inline # ipython의 magic function
value = np.random.standard_normal(40) # 표준 정규분포를 따르는 랜덤값
value
index = range(len(value))
plt.plot(index, value) # 선이나 마커를 플롯하는 함수/ 각각 x,y
# x,y 축 범위 설정
plt.xlim(0,20)
plt.ylim(np.min(value)-1, np.max(value)+1)
plt.figure(figsize = (7,4)) # 그림의 크기(인치) / 튜플값 설정
#numpy.cumsum() : plot의 새로운 figure을 생성
plt.plot(value.cumsum(), 'b', lw=1.5) # 그릴 데이터 - 색(파란), lw = 선 굵기
plt.plot(value.cumsum(), 'ro') # 표시할 모양 - 스타일(빨간동그라미)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Line Plot 1')
```

```python
# 히스토그램 - pyplot.hist() 사용
import matplotlib as mpl
import matplotlib.pyplot as plt # 서브패키지 - 그려주는
import numpy as np

value = np.random.standard_normal((500,2))
plt.figure(figsize=(12,7))
plt.hist(value, label = ['1st', '2nd'], bins = 30, color=('yellow','blue')) # bins 데이터를 구간으로 나누는 개수
plt.grid(True)
plt.legend(bbox_to_anchor = (1.2,1), loc="upper right") # 오른쪽 상단 / 상대적인위치, 기준에서 x,y 이동
plt.xlabel('value')
plt.ylabel('frequency')
plt.title("Histogram 1")
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt # 서브패키지 - 그려주는
import numpy as np

value = np.random.standard_normal(500)
# 모든 패치에 다른 색
cm = plt.cm.get_cmap('Spectral') # 스펙트럴이라는 컬러맵
plt.figure(figsize=(12,6))
n, bins, patches = plt.hist(value, bins=30)
print('1. bins : ' + str(bins))
print("2. the length of bins : " + str(len(bins)))
bin_centers = 0.5 * (bins[:-1] + bins[1:]) # 중앙값
print("3. bin_centers : " + str(bin_centers))
# 최소-최대 정규화 수식 : 새로운 값 = (현-최소)/(최대-최소)
col = (bin_centers - min(bin_centers)) / (max(bin_centers) - min(bin_centers))
print("4. col : " + str(col))

for c, p in zip(col, patches): # zip : 두 개의 변수에 반복자 생성
    plt.setp(p, 'facecolor', cm(c))
plt.xlabel('value')
plt.ylabel('frequency')
plt.title("Histogram 3")
plt.show()
```
